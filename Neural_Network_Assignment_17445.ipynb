```python
import numpy as np
import tensorflow as tf
import matplotlib as mp
```

```python
from tensorflow import keras
%reload_ext tensorboard

import datetime, os
```

```python
import pandas as pd
```

```python
x_test = np.random.rand(10000,10)
```

```python
print(x_test)
```

```python
y = [0]*10000
for i in range(10000):
    for j in range(10):
        y[i] += (j+1)*(x_test[i][j]**(j+1))
#print(y)
y_test = [0]*10000
for i in range(10000):
    y_test[i]= y[i]/30

y_test = np.array(y_test)
print(y_test)
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(10, activation=tf.nn.sigmoid))
model.add(tf.keras.layers.Dense(10, activation=tf.nn.sigmoid))
model.add(tf.keras.layers.Dense(1, activation=tf.keras.activations.linear))

logdir = os.path.join("logs", datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
tensorboard_callback = tf.keras.callbacks.TensorBoard(logdir, histogram_freq=1)
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mean_squared_error'])
%tensorboard --logdir logs

```

```python
model.fit(x_test, y_test, epochs= 300 ,callbacks=[tensorboard_callback])
```

```python
%tensorboard --logdir logs

x_val = np.random.rand(1000,10)
y = [0]*1000
for i in range(1000):
    for j in range(10):
        y[i] += (j+1)*(x_val[i][j]**(j+1))
y_val = np.array(y)
print(y_val)
```

```python
y = [0]*1000
for i in range(1000):
    y[i] = model.predict(np.array([x_val[i]]))*30
y_predict = np.array(y)
print(y)
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 25, 1000)
ax.plot(list(y_val),y_predict.reshape((1000,)));
ax.plot(x,x)
```
